```python
def get_dataframe_info(dataframe):
    # Shape of the DataFrame
    shape = dataframe.shape
    print("Shape:")
    print(shape)
    print()

    # Summary statistics and description of the DataFrame
    description = dataframe.describe()
    print("Description:")
    print(description)
    print()

    # Column names
    column_names = dataframe.columns.tolist()
    print("Column Names:")
    print(column_names)
    print()

    # Data types of the columns
    data_types = dataframe.dtypes
    print("Data Types:")
    print(data_types)
```

```python
# Assuming you have a DataFrame called 'df'
get_dataframe_info(diabetic_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_numerical_columns(dataframe):
    numerical_columns = dataframe.select_dtypes(include=[np.number])

    for column in numerical_columns:
        dataframe[column].plot(kind='hist', title=column)
        plt.show()
```

```python
# Assuming you have a DataFrame called 'diabetic_data'
plot_numerical_columns(diabetic_data)

```
